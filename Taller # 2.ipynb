##**1) Research about the Spectral Clustering method, and answer the following questions:**

##**a) In which cases might it be more useful to apply?**

El cluster spectral puede ser muy útil para el procesamiento de imágenes, ya que su algoritmo se basa en una previa reducción de dimensionalidad a través de los eigen valores y eigen vectores, lo cual lo hace especializado para este tipo de datos, pues su ahorro computacional es grande respecto a otros tipos de clusters. también se usa mucho para redes sociales, pues el algoritmo también tiene una estructura muy definida.

##**b) What are the mathematical fundamentals of it?**

**Matriz de similitud:** se puede utilizar cualquier medida de similitud, pero una medida común es la distancia euclidiana. Si X es una matriz de datos de n x p, entonces la matriz de similitud W de n x n se puede definir como:
W(i,j) = exp(-||X(i,:) - X(j,:)||^2 / 2sigma^2)

**Donde:** ||.|| es la norma euclidiana, X(i,:) es la fila i de X, y sigma es un parámetro que controla la escala de la similitud.

**Grafo:** hay diferentes formas de construir el grafo a partir de la matriz de similitud W. Una opción común es utilizar el grafo completo, donde cada nodo está conectado a todos los demás nodos. Otra opción es utilizar el grafo de vecinos más cercanos, donde cada nodo está conectado a sus k vecinos más cercanos.

**Laplaciano:** una vez que se ha construido el grafo, se puede definir el Laplaciano L de n x n como:

L = D - W

Donde D es una matriz diagonal de n x n con los grados de los nodos en la diagonal, y W es la matriz de similitud. Los grados de los nodos se pueden calcular como la suma de las similitudes de cada nodo con los demás nodos.

Eigen valores y Eigen vectores: los Eigen valores y Eigen vectores del Laplaciano se pueden calcular mediante la siguiente ecuación:
L * v = lambda * v

Donde v es un vector propio (autovector) de tamaño n x 1, y lambda es el correspondiente valor propio (autovalor). Los k vectores propios con los k valores propios más pequeños se utilizan para proyectar los datos en un espacio de menor dimensión.

**K-means:** finalmente, se aplica el algoritmo de K-means en el espacio proyectado para obtener los grupos o clusters finales. El algoritmo de K-means minimiza la suma de las distancias al cuadrado entre cada punto y su centroide asignado.

##**c) What is the algorithm to compute it?**

El algoritmo seria el siguiente: 
1. Definir la matriz de datos X de n x p.
2. Calcular la matriz de similitud W.
3. Construir el grafo a partir de la matriz de similitud W.
4. Calcular el Laplaciano L a partir del grafo.
5. Calcular los k autovectores correspondientes a los k autovalores más pequeños del Laplaciano.
6. Proyectar los datos en un espacio de menor dimensión utilizando los k autovectores.
7. Aplicar el algoritmo de K-means en el espacio proyectado para obtener los clusters finales.
8. Evaluar los clusters mediante alguna medida de calidad, como la pureza o la entropía.

##**2) Research about the DBSCAN method, and answer the following questions:**

###**a) In which cases might it be more useful to apply?**

Este metodo de clusterizacion funciona eligiendo las zonas de mayor densidad, tal cual como lo hace el cerebro humano. Este cluster es útil cuando tenemos formas arbitrarias de los datos. Para resumir, el dbscan funciona de la siguiente manera: DBSCAN sigue la idea de que, para que una observación forme parte de un cluster, tiene que haber un mínimo de observaciones vecinas dentro de un radio de proximidad y de que los clusters están separados por regiones vacías o con pocas observaciones.

###**b) What are the mathematical fundamentals of it?**

Inicialmente mencionare los parámetros fundamentales que recibe el algoritmo DBSCAN, posteriormente mencionare los fundamentos matemáticos: 

**Epsilon (ϵ):** radio que define la región vecina a una observación, también llamada  ϵ-neighborhood

**Minimum points (min_samples):** número mínimo de observaciones dentro de la región épsilon

Respecto a los fundamentos, son los siguientes:

**Densidad:** La densidad se define como el número de puntos en un área determinada. Se expresa como:

ρ(x) = |{x' ∈ D | dist(x, x') ≤ ε}|

**Distancia:** La distancia euclidiana se utiliza comúnmente en DBSCAN y se define como:

dist(x, x') = sqrt(∑i=1n (xi - x'i)^2)

**Vecindario:** El vecindario de un punto x se define como:

Nε(x) = {x' ∈ D | dist(x, x') ≤ ε}

**Puntos centrales:** Un punto central es aquel que tiene al menos un número mínimo de puntos dentro de un radio determinado. Se define como:

corep(x) = {x' ∈ D | dist(x, x') ≤ ε}
|corep(x)| ≥ MinPts

**Alcanzabilidad:** La alcanzabilidad se define en términos de la existencia de una cadena de puntos desde un punto central a otro. Se dice que un punto y es alcanzable desde un punto x si existe una cadena de puntos {p1, p2, ..., pn} tal que p1 = x, pn = y, y cada punto pi+1 está en el vecindario Nε(pi) del punto pi.

**Densidad-conectado:** Dos puntos no centrales están densidad-conectados si existe un punto central que los conecta. Formalmente, si existe un punto x tal que ambos puntos, y e y', son alcanzables desde x

El metodo del codo es una herramienta que sugiere la escogencia optima del numero de clusters a utilizar,el método utiliza la distancia media de las observaciones a su centroide. Es decir, se fija en las distancias intra-cluster.El criterio de eleccion de los clusters se realiza cuando la varianza intra-cluster disminuye significativamente, esto se refleja graficamente como un codo

##**5) Let’s use the newly created modules in unsupervised to cluster some toy data**

####**a) Use the following code snippet to create scattered data X**


```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
import numpy as np

X, y = make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1, center_box=(-10.0, 10.0), shuffle=True, random_state=1)
```

####**b) Plot the resulting dataset. How many clusters are there? How far are they from one another?**




```python
# Perform k-means clustering with 4 clusters
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
kmeans_labels = kmeans.predict(X)

# Perform DBSCAN clustering with eps=1 and min_samples=5
dbscan = DBSCAN(eps=1, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)

# Plot the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels)
plt.title('K-Means Clustering')

plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels)
plt.title('DBSCAN Clustering')

plt.show()
```

Los resultados son bastante distintos, ya las metodologias son completamente diferentes,en kmeans se generan los clusters en base a las distancias de los centroides (distancias intra-clusters e intra-clusters) y con la decision o imposicion del negocio para que sean 4 clusters ,visualmente no me parecen muy buenos, lo cual debemos validar en el siguiente punto (metodo de la silueta), respecto al dbscan este trabaja con densidades entonces se generaron 2 clusters, y se muestran 2 datos atipicos, en lo personal me gusta mas el cluster dbscan pero todo depende del contexto del negocio y su utilidad

####**c) For both k-means and k-medoids (your implementations), calculate the silhouette plots and coefficients for each run, iterating K from 1 to 5 clusters**

```python
from sklearn_extra.cluster import KMedoids
from sklearn.metrics import silhouette_samples, silhouette_score
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmedoids = KMedoids(n_clusters=k, random_state=1)

    kmeans_labels = kmeans.fit_predict(X)
    kmedoids_labels = kmedoids.fit_predict(X)

    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.set_size_inches(18, 7)

    # Silhouette plot for KMeans
    ax1.set_xlim([-0.1, 1])
    ax1.set_ylim([0, len(X) + (k + 1) * 10])
    kmeans_silhouette = silhouette_samples(X, kmeans_labels)
    kmeans_silhouette_avg = silhouette_score(X, kmeans_labels)
    y_lower = 10
    for i in range(k):
        ith_cluster_silhouette_values = kmeans_silhouette[kmeans_labels == i]
        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = cm.nipy_spectral(float(i) / k)
        ax1.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    ax1.set_title("KMeans Silhouette Plot with k=%d" % k)
    ax1.set_xlabel("Silhouette Coefficient Values")
    ax1.set_ylabel("Cluster Label")
    ax1.axvline(x=kmeans_silhouette_avg, color="red", linestyle="--")

    # Silhouette plot for KMedoids
    ax2.set_xlim([-0.1, 1])
    ax2.set_ylim([0, len(X) + (k + 1) * 10])
    kmedoids_silhouette = silhouette_samples(X, kmedoids_labels)
    kmedoids_silhouette_avg = silhouette_score(X, kmedoids_labels)
    y_lower = 10
    for i in range(k):
        ith_cluster_silhouette_values = kmedoids_silhouette[kmedoids_labels == i]
        ith_cluster_silhouette_values.sort()
        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i
        color = cm.nipy_spectral(float(i) / k)
        ax2.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_cluster_silhouette_values,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ax2.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10

    ax2.set_title("KMedoids Silhouette Plot with k=%d" % k)
    ax2.set_xlabel("Silhouette Coefficient Values")
    ax2.set_ylabel("Cluster Label")
    ax2.axvline(x=kmedoids_silhouette_avg, color="red", linestyle="--")

    plt.show()
```

##**6) Use the following code snippet to create different types of scattered data:**



```python
import numpy as np
from sklearn import cluster, datasets, mixture

n_samples = 500
noisy_circles = datasets.make_circles(n_samples=n_samples, factor=0.5, noise=0.05)
noisy_moons = datasets.make_moons(n_samples=n_samples, noise=0.05)
blobs = datasets.make_blobs(n_samples=n_samples, random_state=8)
no_structure = np.random.rand(n_samples, 2), None

####Anisotropically distributed data
random_state = 170
X, y = datasets.make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)
####blobs with varied variances
varied = datasets.make_blobs(
n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state)
```

####**b) Apply k-means, k-medoids, DBSCAN and Spectral Clustering from Scikit-Learn over each dataset and compare the results of each algorithm with respect to each dataset.**

```python
import numpy as np
from sklearn.cluster import KMeans, SpectralClustering, DBSCAN
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt

algorithm_names = ['K-means', 'K-medoids', 'DBSCAN', 'Spectral Clustering']
algorithms = [KMeans(n_clusters=2), KMedoids(n_clusters=2), DBSCAN(eps=0.3, min_samples=5), SpectralClustering(n_clusters=2, affinity='nearest_neighbors')]

datasets = [noisy_circles, noisy_moons, blobs, no_structure, aniso, varied]
dataset_names = ['Noisy Circles', 'Noisy Moons', 'Blobs', 'No Structure', 'Aniso', 'Varied']
for i, dataset in enumerate(datasets):
    X, y = dataset
    fig, axs = plt.subplots(1, len(algorithms), figsize=(15, 5))
    fig.suptitle(dataset_names[i], fontsize=16)
    for j, algorithm in enumerate(algorithms):
        algorithm.fit(X)
        y_pred = algorithm.labels_.astype(int)
        axs[j].scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
        axs[j].set_title(algorithm_names[j])
    plt.show()


```
